```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\satya\Downloads\netflix_titles.csv")
df.head()
```

```python
df = df.fillna(method='ffill') #  Handle missing values
df
```

```python
df = df.drop_duplicates() # df = df.drop_duplicates()
df
```

```python
# Uppercase country and show_id columns
df['COUNTRY'] = df['COUNTRY'].str.upper().replace({'UNITED STATES': 'US', 'SOUTH AFRICA': 'SA'})
df['TYPE'] = df['TYPE'].replace({'Movie': 'M', 'TV Show': 'TS'})
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df
```

```python
# Rename columns to uppercase and replace specific column names
df.columns = df.columns.str.upper()
df
```

```python
print(df.shape)
print(df.info())
print(df.head())
```

```python
print(df.isnull().sum())  # Show nulls per column

# Option 1: Drop rows with nulls
df = df.dropna()
df
```
